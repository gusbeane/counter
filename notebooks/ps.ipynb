### Different encounter angles

```python
ps_iso_fname = basepath + 'anlys/ps/ps_SMUGGLE-iso-lvl4.npy'
ps_iso = np.load(ps_iso_fname, allow_pickle=True).item()
```

```python
Nang = 16

ang_list = np.arange(0, 1, 1/Nang)

ps_list = []
for ang in ang_list:
    ps_fname = basepath + 'anlys/ps/ps_SMUGGLE-flyby-R200-Vphi60-ang16-pro-lvl4-ang'+str(ang)+'.npy'
    ps_list.append(np.load(ps_fname, allow_pickle=True).item())
```

```python
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True, figsize=(12, 10))
ax[0][0].set(xlim=(0, 2), ylim=(0, 60))
plt.subplots_adjust(wspace=0, hspace=0)

for i,x in enumerate(ax.flatten()):
    x.plot(ps_iso['Time'], savgol_filter(ps_iso['PatternSpeed'], 21, 3), c='k')
    
    try:
        ps = ps_list[i]
        x.plot(ps['Time'], savgol_filter(ps['PatternSpeed'], 21, 3))
    except:
        pass
        
# plt.plot(ps_iso['Time'], savgol_filter(ps_iso['PatternSpeed'], 21, 3), c='k')
# plt.ylim(0, 60)
# plt.xlim(0, 2)
```
